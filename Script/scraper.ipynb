```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
from datetime import date , timedelta  


base_url = "https://www.wunderground.com/history/daily/in/new-delhi/VIDD/date/"

options = webdriver.ChromeOptions()
options.add_argument("--headless")

def fetch():
        driver = webdriver.Chrome(options=options)
        data = []
        
        end_window = date.today()
        start_window = end_window - timedelta(days=7)
        
        
        # Loop over this 7-day window
        current_date = start_window
        while current_date <= end_window:
          
            url = f"{base_url}{current_date.year}-{current_date.month:02d}-{current_date.day:02d}"
            print(f"Fetching: {url}")
            driver.get(url)

            time.sleep(5)  

            try:
                tables = driver.find_elements(By.TAG_NAME, "table")
                if len(tables) > 1:
                    table = tables[1]
                    rows = table.find_elements(By.TAG_NAME, "tr")
                    
                    for row in rows[1:]:
                        entry = []
                        columns = row.find_elements(By.TAG_NAME, "td")
                        for column in columns:
                            entry.append(column.text)
                        if entry:
                            entry.append(current_date)  
                            data.append(entry)
            except Exception as e:
                print(f"Error on {current_date}: {e}")
            finally:
                current_date += timedelta(days=1)
        driver.quit()
        return data

def fetch_weather_data():
    data = fetch()
    df = pd.DataFrame(data)
    return df
```
